```python
# Import Dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() aka object-oriented method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create quick graph using object-oriented interface method
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create scatter plot using the plt.plot() function.
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create scatter plot with labels and color. Skilldrill 5.1.7a.
plt.scatter(y_axis, x_axis, s=50, color="red", label='Chicago')
plt.gca().invert_yaxis()
plt.xlim(0, 45)
plt.xlabel ('Fare($)')
plt.ylabel ('Date')
plt.title ('Pyber Fare by Month')
# Add the legend.
plt.legend()
```

```python
plt.scatter(x_axis, y_axis)
```

```python
# Change size of marker by using size paarmeter 's'.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Adjust size by multiplying the data in the y-axis by any value. 

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# List Comprehension can replace for and while loops
# format is: new_list = [expression for item in list if conditional]
```

```python
# Use list comprehension to multiple data point in the y_axis list. 
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

```python
# Create a scatter plot using object-oriented approach
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a bubble chart using Object-Oriented Interface
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Create bubble chart with object-oriented approach Skilldrill 5.1.7b
fig, ax = plt.subplots() 
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color="skyblue",label='Boston', alpha=.80, linewidth=2, edgecolor="k")
ax.invert_yaxis()
# Create labels for the x and y axes.
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
# Set the y limit between 0 and 45.
ax.set_xlim(0, 45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add the legend.
ax.legend()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
